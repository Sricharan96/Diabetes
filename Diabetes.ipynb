```python
import pandas as pd
pd.read_csv('diabetes.csv')
diabetes_df=pd.read_csv('diabetes.csv')
diabetes_df
diabetes_df.shape

y=diabetes_df['Outcome']
y

diabetes_df.drop(columns='Outcome',inplace=True)
diabetes_df

#train test split

import sklearn 
from sklearn.model_selection import train_test_split
train_test_split(diabetes_df,y,test_size=0.2)
len(train_test_split(diabetes_df,y,test_size=0.2))

from sklearn.linear_model import LogisticRegression
X_train,X_test,y_train,y_test=train_test_split(diabetes_df,y,test_size=0.2,random_state=42)
logreg=LogisticRegression()
logreg.fit(X_train,y_train)

import warnings 
warnings.filterwarnings('ignore')

logreg.fit(X_train,y_train)
logreg.predict(X_test)
len(logreg.predict(X_test))
len(X_test)
y_test_pred=logreg.predict(X_test)
y_test
y_test_pred
len(y_test_pred)
logreg.predict(X_train)
y_train_pred=logreg.predict(X_train)

from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,mean_squared_error
pd.DataFrame({'test_actuals':y_test,'test_pred':y_test_pred})
confusion_matrix(y_test,y_test_pred)
precision_score(y_test,y_test_pred)
recall_score(y_test,y_test_pred)
f1_score(y_test,y_test_pred)

logreg.predict_proba(X_test)
y_test_pred

probabs=logreg.predict_proba(X_test)
probabs
logreg.predict_proba(X_test)[::-1]
pos_probabs=logreg.predict_proba(X_test)[::,1]
pos_probabs


#increasing and decreasing threshold graph program

from sklearn.metrics import roc_curve
roc_curve(y_test,pos_probabs)
fpr,tpr,thresholds=roc_curve(y_test,pos_probabs)
fpr,tpr,thresholds

from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,pos_probabs)

# To draw Curve(AUC)

import matplotlib.pyplot as plt
plt.plot(fpr,tpr)

#to label the graph

plt.plot(fpr,tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUC Curve')
```
